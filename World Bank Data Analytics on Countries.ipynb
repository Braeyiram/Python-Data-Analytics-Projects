## Machine Learning Integeration


### 1: Data Preparation

### Flagging Missing Values

```python
# Add binary flags for missing data
for column in new_data.columns:
    if new_data[column].isnull().any():
        new_data[f'Missing_{column}'] = new_data[column].isnull().astype(int)
new_data.head()
```

### Handling Missing Values in Target

Since the **HistGradientBoostingRegressor** require the target variable (y) to have no missing values, I will train a machine learning model (**Random Forest Regressor**) to predict missing values for the target variable `(Net Trade)` based on other features in the dataset.


#### Data Splitting
1. **Training data** (`train_data`) to teach the model patterns in the data.
2. **Validation data** to evaluate the model's performance and adjust its parameters.
3. **Test data** is used for predictions, filling in missing values in the `Net Trade` column.

```python
# Split data into rows with and without missing Net Trade values
train_data = new_data.dropna(subset=['Net Trade'])  # Rows with valid Net Trade
test_data = new_data[new_data['Net Trade'].isna()]  # Rows with missing Net Trade
```

```python
# Define training features (X) and target (y)
X_train_rf = train_data.drop(columns=['Net Trade'])  # Drop the target column
y_train_rf = train_data['Net Trade']  # Target column

# Define testing features
X_test_rf = test_data.drop(columns=['Net Trade'])  # Only features
```

```python
# Split train_data into training and validation sets
X_train_rf_split, X_val_rf, y_train_rf_split, y_val_rf = train_test_split(
    X_train_rf, y_train_rf, test_size=0.2, random_state=42)
```

#### 2. Handling Categorical Data
Using **OneHotEncoding** to convert them into numerical values the model can understand.


```python
# Combine train_data and test_data before encoding
new_data_combined = pd.concat([train_data, test_data])

# Handle categorical columns in the combined dataset
categorical_cols = new_data_combined.select_dtypes(include=['object', 'category']).columns.tolist()

if categorical_cols:
    # Fit the encoder on the combined dataset
    encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
    encoder.fit(new_data_combined[categorical_cols])

    # Transform categorical columns in the combined dataset
    encoded_cats_combined = encoder.transform(new_data_combined[categorical_cols])
    encoded_df_combined = pd.DataFrame(
        encoded_cats_combined,
        index=new_data_combined.index,
        columns=encoder.get_feature_names_out(categorical_cols)
    )

    # Drop original categorical columns and add encoded ones
    new_data_encoded = new_data_combined.drop(columns=categorical_cols).join(encoded_df_combined)

else:
    # If no categorical columns, proceed without encoding
    new_data_encoded = new_data_combined.copy()
```

#### 3. Training Random Forest Regressor

```python
#  Random Forest Imputation

# Define training and testing datasets from the encoded data
X_train_rf_encoded = new_data_encoded.loc[train_data.index].drop(columns=['Net Trade'])  # Training features
y_train_rf = train_data['Net Trade']  # Training target
X_test_rf_encoded = new_data_encoded.loc[test_data.index].drop(columns=['Net Trade'])  # Testing features

# Initialize and train the Random Forest model
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train_rf_encoded, y_train_rf)

# Predict missing values in the test dataset
test_data['Net Trade'] = rf.predict(X_test_rf_encoded)

# Combine the filled datasets
new_data_filled = pd.concat([train_data, test_data])
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Predict on the training set
rf_predictions_train = rf.predict(X_train_rf_encoded)

# Calculate metrics
rf_mse_train = mean_squared_error(y_train_rf, rf_predictions_train)
rf_r2_train = r2_score(y_train_rf, rf_predictions_train)
rf_mae_train = mean_absolute_error(y_train_rf, rf_predictions_train)

# Print metrics
print(f"Training MSE: {rf_mse_train:.2f}")
print(f"Training R²: {rf_r2_train:.2f}")
print(f"Training MAE: {rf_mae_train:.2f}")
```

#### 4. Hyperparameter Tuning

The Random Forest model has parameters like `max_depth`, `n_estimators`, and others that can greatly impact its performance. Instead of manually trying out different values, we use **GridSearchCV** to evaluate multiple combinations automatically and identify the best configuration based on the R² scoring metric.

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    'max_depth': [5, 10, 15, None],
    'n_estimators': [100, 200, 300],
    'max_features': ['sqrt', 'log2', None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='neg_mean_squared_error',
    verbose=2,
    n_jobs=-1  # Use all processors for parallel computation
)

# Fit GridSearchCV on the training data
grid_search.fit(X_train_rf_encoded, y_train_rf)

# Print the best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best MSE:", -grid_search.best_score_)
```

```python
best_rf = RandomForestRegressor(
    max_depth=15,
    max_features='log2',
    min_samples_leaf=2,
    min_samples_split=2,
    n_estimators=100,
    random_state=42
)
best_rf.fit(X_train_rf_encoded, y_train_rf)
```

```python
# Re-impute missing values using the optimized model
test_data['Net Trade'] = best_rf.predict(X_test_rf_encoded)

# Combine imputed test data with training data
imputed_data_rf = pd.concat([train_data, test_data]).sort_index()
```

### 2: Define Features and Target
Now that the missing values in `Net Trade` have been filled, we can use the complete dataset to train the main model (**HistGradientBoostingRegressor**) for the analysis.


```python
# Define features (X) and target (y)
X_hgb = imputed_data_rf.drop(columns=['Net Trade'])  # All features
y_hgb = imputed_data_rf['Net Trade']                # Target column
```

```python
# Train the model on the entire dataset
X_train_hgb = X_hgb
y_train_hgb = y_hgb
```

### 3: Train the Main Machine Learning Model

```python
# Identify columns with non-numeric data
non_numeric_cols = X_hgb.select_dtypes(include=['object', 'category']).columns
print("Non-numeric columns:", non_numeric_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Apply the encoder on the non-numeric columns
encoded_cats = encoder.fit_transform(X_hgb[['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']])

# Create a DataFrame with encoded columns
encoded_df = pd.DataFrame(
    encoded_cats,
    columns=encoder.get_feature_names_out(['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']),
    index=X_hgb.index
)

# Drop original non-numeric columns and add the encoded ones
X_hgb_encoded = X_hgb.drop(columns=['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']).join(encoded_df)
```

```python
print(X_hgb_encoded.dtypes)
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

# Initialize and train the model
hgb_model = HistGradientBoostingRegressor(
    max_iter=100,
    learning_rate=0.1,
    max_depth=None,
    l2_regularization=1.0,
    random_state=42
)

hgb_model.fit(X_hgb_encoded, y_hgb)
```

```python
# Use the same encoder for prediction
encoded_cats_pred = encoder.transform(X_hgb[['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']])

# Create a DataFrame for prediction
encoded_df_pred = pd.DataFrame(
    encoded_cats_pred,
    columns=encoder.get_feature_names_out(['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']),
    index=X_hgb.index
)

# Align the prediction dataset
X_hgb_encoded_pred = X_hgb.drop(columns=['Country Name', 'Country Code', 'Continent Name', 'Income Level', 'Trade Category']).join(encoded_df_pred)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Predict on the training set
hgb_predictions_train = hgb_model.predict(X_hgb_encoded_pred)

# Calculate metrics
hgb_r2_train = r2_score(y_train_hgb, hgb_predictions_train)
hgb_mse_train = mean_squared_error(y_train_hgb, hgb_predictions_train)
hgb_mae_train = mean_absolute_error(y_train_hgb, hgb_predictions_train)

# Print the metrics
print(f"HistGradientBoosting Training R²: {hgb_r2_train:.2f}")
print(f"HistGradientBoosting Training MSE: {hgb_mse_train:.2f}")
print(f"HistGradientBoosting Training MAE: {hgb_mae_train:.2f}")
```
